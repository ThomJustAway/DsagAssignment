### a. Briefly describe two main differences between Big O notation and Big Omega  notation. (6 marks)

The big O notation find the worst possible time it takes to run a function. However, The Big Omega find the best possible time it takes to run a function. For instance, the algorithm runs on a function called T(N). The big O notation will have a function that is either be more or equal to the original function T(N). The opposite goes for Big Omega notation where it will have function the will be either less or equal to the original function T(N).

Another difference how the functions that satisfy the Big O notation mostly does not satisfy the Big Omega notation. Let say the function T(N) is tested where it runs on linear time. function like n log(N) , n^n and so on will satisfy the Big O notation of set function. However, putting this function in the Big Omega notation will not work as the function will soon pass T(N). It is vice versa for the Big Omega function.The only other function that satisfy both condition will be 'N' Where it is linear time and will work for both Big Omega and Big O notation .

Despite the quick sort having a worst big o notation than the merge sort, some people will stick to quick sort as the normal possible case is a big O notation of O(n log(n)) and the quick sort alogrithm is actually faster compared to the merge sort in a normal case. ### A program is required to record and display the results of a swimming competition. The swimming pool has 5 lanes. The swimmer name for each lane is first recorded. After the race, the timings for the swimmers are captured. The results for the race is then displayed.


### a.	Write a function inputSwimmers() that prompts for 5 swimmers' names and returns the names in a list. An example of the input session is as follows: ( 3 marks)
#### Enter lane 1 swimmer: Jin Jia Ho
#### Enter lane 2 swimmer: Mai Siao Siao
#### …
#### Enter lane 5 swimmer: Ho Seh Lah



```python
def inputSwimmer():
    array = []
    for i in range(1,6): 
        swimmer = input("Enter lane "+str(i)+" swimmer:") #prompting the user to give a name
        array.append(swimmer) #putting the user inside of the list
    return array
```

### b.	Write another function inputTiming(swimmers) that has the list of swimmers names as paramter. The function creates an empty list, prompts and stores the timing for each of the swimmers. The timing list is returned. An example of the input session is as follows: (3 marks)
#### Enter timing for Jin Jia Ho: 47.15
#### Enter timing for Mai Siao Siao: 46.91
#### ….
#### Enter timing for Ho Seh Lah: 48.01


```python
def inputTiming(swimmers):
    array = [] 
    for swimmer in swimmers:
        timing = float(input("Enter timing for "+swimmer+":")) # prompt the user for the timing of the swimmer
        array.append(timing)
    return array
  
```

### c.	Write another function printResults(swimmers, timings) that has the name and timing lists as parameters. The function prints a summary of the results as follows: (3 marks)
#### Jin Jia Ho           47.15s
#### Mai Siao Siao     46.91s
#### …
#### Ho Seh Lah        48.01s
#### The Fastest timing is 46.91s


```python
def printResults(swimmers,timings):
    lowestTiming=timings[0]
    for i in range(5): #repeating this 5 times
        print(f"{swimmers[i]:<20s} {timings[i]}") #printing out the swimmer name and time 
        if (timings[i]<lowestTiming):
            lowestTiming=timings[i] # changing the value of lowest if a new lower value is found
        
    
    print("The Fastest timing is "+str(lowestTiming)+"s") #printing the result out. Timings[0] is the fastest timing
```

### d.	Modify the function in part (c) and name it printResults2(swimmers, timings) that prints the summary in sorted ascending order of the timing. Therefore, the summary may look like this: (5 marks)
#### Mai Siao Siao	46.91s
#### Jin Jia Ho	 47.15s
#### …
#### Ho Seh Lah	48.01s

```python
def printResults2(swimmers,timings):
    arrayOfSwimmer=[]
    for i in range(len(swimmers)): 
        swimmer = [swimmers[i],timings[i]] # storing the swimmer name and timing inside the swimmer variable
        arrayOfSwimmer.append(swimmer)
    
    #defining the sorting method
    def sorter(e):
        return e[1]
    arrayOfSwimmer.sort(key = sorter) # sort the list of swimmer from fastest to slowest
    
    for swimmer in arrayOfSwimmer:
        print(f'{swimmer[0]:<20s}{swimmer[1]}s') 
```
